```python
# Categorical boolean mask
categorical_feature_encoding = df_c2.dtypes==object
# filter categorical columns using mask and turn it into a list
categorical_cols = df_c2.columns[categorical_feature_encoding].tolist()
categorical_cols
```

```python
# import labelencoder
from sklearn.preprocessing import LabelEncoder
# instantiate labelencoder object
le = LabelEncoder()
```

```python
# apply le on categorical feature columns
df_c2[categorical_cols] = df_c2[categorical_cols].apply(lambda col: le.fit_transform(col))
df_c2[categorical_cols].head(10)
```

```python
from surprise import Dataset,Reader
from surprise.model_selection import cross_validate
from surprise import NormalPredictor
reader = Reader(rating_scale=(1, 10))
data1 = Dataset.load_from_df(df_c2[['product', 'author', 'score']], reader)
```

```python
from surprise.model_selection import train_test_split
trainset, testset = train_test_split(data1, test_size=.25,random_state=123)
```

```python
trainset.all_ratings()
```

```python
print(trainset.to_raw_uid(0))
```

```python
from surprise import SVD, KNNWithMeans
from surprise import accuracy
```

```python
svd_model = SVD(n_factors=5,biased=False)
svd_model.fit(trainset)
```

```python
testset[0]
```

```python
test_pred = svd_model.test(testset)
```

# 5. Evaluate the collaborative model. Print RMSE value. 

```python
accuracy.rmse(test_pred)
```

```python
from surprise import KNNWithMeans
from surprise import accuracy

itemb = KNNWithMeans(k=10, sim_options={ 'user_based': False})

itemb.fit(trainset)
```

- 5. Evaluate the collaborative model. Print RMSE value. 

```python
test_pred=itemb.test(testset)
print(accuracy.rmse(test_pred))
```

```python
from surprise import KNNWithMeans
from surprise import accuracy

userb = KNNWithMeans(k=10, sim_options={ 'user_based': True})

userb.fit(trainset)
```

- 5. Evaluate the collaborative model. Print RMSE value. 

```python
test_pred=userb.test(testset)
print(accuracy.rmse(test_pred))
```
